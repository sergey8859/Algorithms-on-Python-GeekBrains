```python
import pandas as pd
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
df = pd.DataFrame({'whoAmI':lst})
df1 = pd.DataFrame()
df.head(20)
```

```python
def wh(x):
  if x == 'human':
    return 1
  else:
    return 0
def wh1(x):
  if x == 'robot':
    return 1
  else:
    return 0  


 
```

```python
df1['human'] = df.whoAmI.apply(wh)
df1['robot'] = df.whoAmI.apply(wh1)
```
