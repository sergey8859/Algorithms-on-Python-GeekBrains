```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('sample_data/california_housing_train.csv')
```

```python
sns.scatterplot(data=df, x = df.query("households <=  4500")['households'], y = df.query("population <= 15000")['population'])
```

```python
sns.relplot(x = df.longitude, y = df.median_house_value, kind = 'line')
```

```python
sns.histplot(data = df, x = 'housing_median_age')
```

```python
sns.displot(df, x="median_house_value", hue="housing_median_age")
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
sns.scatterplot(x = penguins.bill_length_mm, y = penguins.bill_depth_mm, hue = penguins.sex, size = 5)
```

```python
sns.scatterplot(x = penguins.flipper_length_mm, y = penguins.body_mass_g, hue = penguins.island, size = 5)
```

```python
cols = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'sex']
g = sns.PairGrid(penguins[cols], hue = 'sex')
g.map(sns.scatterplot)
```

```python
# Создать новый столбец в таблице с пингвинами, который будет отвечать за показатель длины клюва пингвина.
# high - длинный(от 42)
# middle - средний(от 35 до 42)
# low - маленький(до 35)
def length(x):
  if x >= 42:
    return 'hing'
  elif x > 35 and x < 42:
    return 'middle'
  elif x <= 35:
    return 'low'
penguins['length'] = penguins.bill_length_mm.apply(length)
penguins
```

```python
sns.displot(x = penguins.flipper_length_mm, hue=penguins.length)
```
